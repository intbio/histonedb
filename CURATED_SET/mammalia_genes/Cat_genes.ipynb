### Get protein GeneBank records via `ncbi_id` from RefSeq

**accessions are retrived only for records from RefSeq**

```python
dict_accessions, dict_sequences = {}, {}
add_rows = []

for gene_id in genes_vgnc_merged[~genes_vgnc_merged["ncbi_id"].isna()][
    "ncbi_id"
].unique():
    print(gene_id)
    with Entrez.elink(
        dbfrom="gene", db="protein", id=str(gene_id), term="refseq[filter]"
    ) as handle:
        link_results = Entrez.read(handle)
    # Получить Protein IDs
    protein_ids = []
    if link_results[0]["LinkSetDb"]:
        protein_ids = [link["Id"] for link in link_results[0]["LinkSetDb"][0]["Link"]]
    # print("Количество связанныех Protein IDs:", len(protein_ids))
    if protein_ids:
        with Entrez.efetch(
            db="protein", id=",".join(protein_ids), rettype="gb", retmode="text"
        ) as handle:
            protein_records = list(SeqIO.parse(handle, "genbank"))
        # print("Количество RefSeq proteins:", len(protein_records))
        # Сохраним в словарь найденные accession и последовательность
        dict_accessions[gene_id] = protein_records[0].id
        dict_sequences[protein_records[0].id] = str(protein_records[0].seq)
        # Если последовательностей больше одной, то сгенерируем словари для того, чтобы добавить их в таблицу
        if len(protein_records) > 1:
            print(
                ", ".join([protein_records[i].id for i in range(len(protein_records))])
            )
            for pr in protein_records[1:]:
                row_asdict = (
                    genes_vgnc_merged[genes_vgnc_merged["ncbi_id"] == gene_id]
                    .iloc[0]
                    .to_dict()
                )  # Находим сроку с ncbi_id равным gene_id
                row_asdict.update({"accession": pr.id, "sequence": str(pr.seq)})
                add_rows.append(row_asdict)
```

```python
pd.DataFrame(add_rows)
```

```python
genes_vgnc_merged["accession"] = genes_vgnc_merged["ncbi_id"].map(dict_accessions)
genes_vgnc_merged["sequence"] = genes_vgnc_merged["accession"].map(dict_sequences)
genes_vgnc_merged.head()
```

```python
genes_vgnc_merged.shape
```

```python
genes_vgnc_merged = pd.concat([genes_vgnc_merged, pd.DataFrame(add_rows)])
print(genes_vgnc_merged.shape)
genes_vgnc_merged.head()
```
