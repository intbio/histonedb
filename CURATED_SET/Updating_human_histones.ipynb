```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/intbio/hist_gene_prot/master/docs/human_histones.csv?token=GHSAT0AAAAAAB32GIFDIRWV3UE7YRNO4SBCY55W5JQ",
                 engine='python').fillna('')
df.columns = ['type', 'Histone variant', 'Clustered (canonical) isoform',
              'hgnc_gene_name', 'ncbi_gene_id', 'Ensembl gene ID',
              'Transcript stable ID', 'RefSeq mRNA ID', 'RefSeq peptide ID',
              'Protein length', 'references']
df.shape, df.columns
```

```python
df['accession'] = df["RefSeq peptide ID"]
df['variant_group'] = ['']*df.shape[0]
df['variant'] = df["Histone variant"]
df['taxonomy_id'] = ['9606']*df.shape[0]
df['organism'] = ['Homo sapiens']*df.shape[0]
df['phylum'] = ['Chordata']*df.shape[0]
df['class'] = ['Mammalia']*df.shape[0]
df['taxonomy_group'] = ['Mammalia']*df.shape[0]
df['info'] = ['']*df.shape[0]
df = df[['accession', 'type', 'variant_group', 'variant',
         'Histone variant', 'Clustered (canonical) isoform',
         'ncbi_gene_id', 'hgnc_gene_name',
         'Ensembl gene ID', 'Transcript stable ID', 'RefSeq mRNA ID', 'RefSeq peptide ID', 'Protein length',
         'taxonomy_id', 'organism', 'phylum', 'class', 'taxonomy_group', 'info', 'references']]
df.index = df['accession']
df.head()
```

```python
df['Protein length'] = [int(l) if l != '' else '' for l in df['Protein length']]
df.head()
```
