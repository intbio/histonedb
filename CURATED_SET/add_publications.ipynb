```python
import json, re
```

```python
with open('classification.json') as json_file:
    data = json.load(json_file)
```

```python
for _, d in data['info'].items():
    print(d['publications'])
```

```python
for _, d in data['info'].items():
    try:
        for art in re.findall('\[(.*?)\]', d['description']['summary']):
            for a in art.split(','):
                try:
                    if a not in d['publications']: d['publications'].append(a)
                except KeyError: d['publications'] = [a]
    except TypeError: continue
```

```python
for _, d in data['info'].items():
    print(d['publications'])
```

```python
with open('classification_2.json', 'w') as json_file:
    json.dump(data, json_file)
```
