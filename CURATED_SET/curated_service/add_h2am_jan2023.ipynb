```python
df = pd.read_csv('curated_service/h2am_histones.csv')
df.index = df.accession
df = df[['accession', 'Species', 'Sequence']]
df.shape, df.columns
```

```python
df.columns = ['accession', 'organism', 'sequence']
df['type'] = ['H2A']*df.shape[0]
df['variant_group'] = ['H2A.M']*df.shape[0]
df['variant'] = ['H2A.M']*df.shape[0]
df['references'] = ['25983206']*df.shape[0]
df = df[['accession', 'type', 'variant_group', 'variant', 'organism', 'references', 'sequence']]
df.head()
```

```python
cs.data = pd.concat([cs.data, df]).fillna('')
cs.data.shape, list(cs.has_duplicates())
```

```python
cs.update_accession_version()
cs.data=cs.data.set_index(cs.data.accession.values)
```
