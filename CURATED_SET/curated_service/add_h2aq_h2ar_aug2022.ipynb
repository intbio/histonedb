```python
new_seqs = []
for s in SeqIO.parse('/home/l_singh/_scratch/H2A/h2aq.fasta', 'fasta'):
    taxid = ''
    for d in s.description.split(' '):
        if 'taxid' in d: taxid = d.split('=')[1]
    if taxid=='None': taxid = ''
    new_seqs.append({'accession': s.id, 'type': 'H2A', 'variant_group': 'H2A.Q', 'variant': '', 'doublet': False, 'gi': '',
                     'ncbi_gene_id': '', 'hgnc_gene_name': '', 'taxonomy_id': taxid, 'organism': s.description.split('[')[1].split(']')[0], 'phylum': '',
                     'class': '', 'taxonomy_group': '', 'info': '', 'references': '29549088',
                     'sequence': str(s.seq)})
for s in SeqIO.parse('/home/l_singh/_scratch/H2A/h2ar.fasta', 'fasta'):
    taxid = ''
    for d in s.description.split(' '):
        if 'taxid' in d: taxid = d.split('=')[1]
    if taxid=='None': taxid = ''
    new_seqs.append({'accession': s.id, 'type': 'H2A', 'variant_group': 'H2A.R', 'variant': '', 'doublet': False, 'gi': '',
                     'ncbi_gene_id': '', 'hgnc_gene_name': '', 'taxonomy_id': taxid, 'organism': s.description.split('[')[1].split(']')[0], 'phylum': '',
                     'class': '', 'taxonomy_group': '', 'info': '', 'references': '29549088',
                     'sequence': str(s.seq)})
df = pd.DataFrame(new_seqs)
df
```

```python
cs.data = pd.concat([cs.data, df])
```
