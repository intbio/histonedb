### Get protein GeneBank records via `ncbi_id` from RefSeq

**accessions are retrived only for records from RefSeq**

```python
dict_accessions, dict_sequences = {}, {}
add_rows = []

for gene_id in genes_vgnc_merged[~genes_vgnc_merged["ncbi_id"].isna()][
    "ncbi_id"
].unique():
    print(gene_id)
    with Entrez.elink(
        dbfrom="gene", db="protein", id=str(gene_id), term="refseq[filter]"
    ) as handle:
        link_results = Entrez.read(handle)
    # Получить Protein IDs
    protein_ids = []
    if link_results[0]["LinkSetDb"]:
        protein_ids = [link["Id"] for link in link_results[0]["LinkSetDb"][0]["Link"]]
    # print("Количество связанныех Protein IDs:", len(protein_ids))
    if protein_ids:
        with Entrez.efetch(
            db="protein", id=",".join(protein_ids), rettype="gb", retmode="text"
        ) as handle:
            protein_records = list(SeqIO.parse(handle, "genbank"))
        # print("Количество RefSeq proteins:", len(protein_records))
        # Сохраним в словарь найденные accession и последовательность
        dict_accessions[gene_id] = protein_records[0].id
        dict_sequences[protein_records[0].id] = str(protein_records[0].seq)
        # Если последовательностей больше одной, то сгенерируем словари для того, чтобы добавить их в таблицу
        if len(protein_records) > 1:
            print(
                ", ".join([protein_records[i].id for i in range(len(protein_records))])
            )
            for pr in protein_records[1:]:
                row_asdict = (
                    genes_vgnc_merged[genes_vgnc_merged["ncbi_id"] == gene_id]
                    .iloc[0]
                    .to_dict()
                )  # Находим сроку с ncbi_id равным gene_id
                row_asdict.update({"accession": pr.id, "sequence": str(pr.seq)})
                add_rows.append(row_asdict)
```

```python
pd.DataFrame(add_rows)
```

```python
genes_vgnc_merged["accession"] = genes_vgnc_merged["ncbi_id"].map(dict_accessions)
genes_vgnc_merged["sequence"] = genes_vgnc_merged["accession"].map(dict_sequences)
genes_vgnc_merged.head()
```

```python
genes_vgnc_merged.shape
```

```python
genes_vgnc_merged = pd.concat([genes_vgnc_merged, pd.DataFrame(add_rows)])
print(genes_vgnc_merged.shape)
genes_vgnc_merged.head()
```

```python
import requests
```

```python
def get_protein_record(uniprot_id):
    url = f"https://www.uniprot.org/uniprot/{uniprot_id}.xml"
    response = requests.get(url)

    if response.ok:
        # Парсим XML (можно использовать BeautifulSoup или Bio.UniProt)
        ref_seq_reference = response.text.split('<dbReference type="RefSeq" id="')
        if len(ref_seq_reference) < 2:
            print(f"UniProt ID {uniprot_id} не найден")
            return
        refseq_id = response.text.split('<dbReference type="RefSeq" id="')[1].split(
            '"'
        )[0]

        # Шаг 2: Получаем GenBank запись
        # with Entrez.efetch(
        #     db="protein", id=refseq_id, rettype="gb", retmode="text"
        # ) as handle:
        #     protein_record = SeqIO.read(handle, "genbank")
        # print(protein_record)
        return refseq_id
    print(f"UniProt ID {uniprot_id} не найден")


get_protein_record("A0A5S6RBZ5")
```

```python
dict_acc_uni_str, dict_seq_uni_str = {}, {}
for uniprot_str in genes_vgnc_merged[
    genes_vgnc_merged["accession"].isna() & ~(genes_vgnc_merged["uniprot_ids"].isna())
]["uniprot_ids"].unique():
    print(uniprot_str)
    acc_list = []
    for uid in uniprot_str.split("|"):
        refseq_id = get_protein_record(uid)
        if refseq_id and refseq_id not in acc_list:
            acc_list.append(refseq_id)
    dict_acc_uni_str[uniprot_str] = "|".join(acc_list)
dict_acc_uni_str
```

Оказалось, что в человеческом геноме есть **однозначная** позиция для начала транскрипции (которая соотвтетствует кодону ATG), а у **шимпанзе** есть **две** позиции: обе соответствуют стартовому кодону **ATG**, но первая позиция, с которой начинается предсказание аминокислотной последовательности лежит левее, поэтому на выходе получилась последовательность, которая с 18-й аминокислоты **совпадает** с гомологичным белком из человека.
